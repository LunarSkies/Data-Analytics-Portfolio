### Cleaning the Wikipedia Data 

```python
table
```

```python
# find any four digits at the beginning of a string for our Founded column and replace it
table['Founded'] = table['Founded'].str.extract(r'^(\d{4})')
table
```

```python
# Replace GICS Sector with just Sector
table.rename({'GICS Sector': 'Sector'}, axis=1, inplace=True)
table
```

```python
# Replace GICS Sub-Sector with just Sub-Sector
table.rename({'GICS Sub-Industry': 'Sub-Industry'}, axis=1, inplace=True)
table
```

```python
# Splitting HQ Location at the ,
table[['City', 'State','Null']] = table['Headquarters Location'].str.split(',', expand=True)
table
```

```python
# Droping unimportant columns
clean_table = table.drop(['Null','CIK','Date first added','Headquarters Location','SEC filings'], axis=1)
clean_table
```

```python
# Save Clean file
clean_table.to_csv('clean_table.csv')
```

### Join the two tables

```python
df1 = pd.read_csv('clean_table.csv')
df2 = pd.read_csv('new_data.csv')
final_table = df1.merge(df2, how='left', on='Symbol')
final_table
```

```python
# Drop unimportant columns
final_table = final_table.drop(['Unnamed: 0_y','Unnamed: 0_x'], axis=1)
final_table
```

```python
# Remove nulls and duplicates
final_table = final_table.dropna().drop_duplicates()
final_table
```
